```python
#Replace unnecessary info (like the word 'civil', '400th judicial district court', etc.)
new_body = body.replace('\n\nTXFTBENDPROD','').replace('\n\nCivil','').replace('\n\n400th Judicial\nDistrict Court','')
new_body
```

```python
#Use regex to replace the date pieces
new_body = re.sub(r'\n\nDate Range: [0-9]{2}/[0-9]{2}/[0-9]{4} to [0-9]{2}/[0-9]{2}/[0-9]{4}', '', new_body)
new_body = re.sub(r'\n\nPrinted on [0-9]{2}/[0-9]{2}/[0-9]{4} [0-9]{1,2}:[0-9]{1,2} [AMP]{2}\n\nPage\n\n[0-9]{1,3}\n\nof\n\n[0-9]{1,3}', '', new_body)
new_body
```

```python
#Now find the number of cases on this page
num_cases = len(re.findall(r'[0-9]{2}-[DCVR]{3}-[0-9A-Z]*', new_body))
num_cases
```

```python
#Is there any order to this info?
#Split on the double new lines and see what that list looks like
body_parts = new_body.split('\n\n')
body_parts
```

```python
#Remove header info
body_parts.remove('Pending Cases\nCase Categories:')
body_parts.remove('Case Types: All')
body_parts.remove('Locations: 400th Judicial District Court')
body_parts.remove('Primary Sort:')
body_parts.remove('Case Number')
body_parts.remove('Case Number')
body_parts.remove('Date Range Type:')
body_parts.remove('Case Filed Date')
body_parts.remove('Location')
body_parts.remove('Judicial Officer')
body_parts.remove('Case\nCategory')
body_parts.remove('Case Type\nCase Subtype')
body_parts.remove('Case Status')
body_parts.remove('Style')
```
