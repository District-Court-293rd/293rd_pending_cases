```python
def parse_civil_report(content):
    """
    This function takes in the text of the PDF report and generates a df.
    The df will still need to be cleaned up, and other static columns added.
    """
    
    #Split into individual pages
    pages = content.strip().split('\n\n\x0c')
    
    #How many pages?
    num_pages = len(pages)
    print(f"There are {num_pages} pages.")
    
    #Provide a list of possible statuses in order to identify the status data piece
    possible_statuses = [
        'Active',
        'Inactive',
        'Pending',
        'Re-Opened'
    ]
    
    possible_types = [
        'Contract -\nConsumer/Commercial/Debt',
        'Injury or Damage - Other Injury\nor Damage',
        'Injury or Damage Involving Motor\nVehicle',
        'Other Civil',
        'Other Tax',
        'Injury or Damage Medical\nMalpractice',
        'Real Property - Other Real\nProperty',
        'Contract - Other Contract',
        'Seizure and Forfeiture',
        'Judgment NISI',
        'Filing of Fraudulent Lien - Civil'
    ]
    
    #Create an empty df. We will use this one to concat each page's dataframe to
    df = pd.DataFrame()
    
    #Iterate through each page
    for page_num, page in enumerate(pages):
        
        #Reset my_dict
        my_dict = {}
        
        #Replace unnecessary info (like the word 'civil', '400th judicial district court', etc.)
        page = page.replace('\n\nTXFTBENDPROD','').replace('\n\nCivil','').replace('\n\n400th Judicial\nDistrict Court','')
        
        #Use regex to replace the date pieces
        page = re.sub(r'\n\nDate Range: [0-9]{2}/[0-9]{2}/[0-9]{4} to [0-9]{2}/[0-9]{2}/[0-9]{4}', '', page)
        page = re.sub(r'\n\nPrinted on [0-9]{2}/[0-9]{2}/[0-9]{4} [0-9]{1,2}:[0-9]{1,2} [AMP]{2}\n\nPage\n\n[0-9]{1,3}\n\nof\n\n[0-9]{1,3}', '', page)
        
        #Now find the number of cases on this page
        num_cases = len(re.findall(r'[0-9]{2}-[DCVR]{3}-[0-9A-Z]*', page))
        print(f'Page {page_num + 1} has {num_cases} cases.')
        
        #Split on the double new lines
        data_pieces = page.split('\n\n')
        
        #Remove header info
        data_pieces.remove('Pending Cases\nCase Categories:')
        data_pieces.remove('Case Types: All')
        data_pieces.remove('Locations: 400th Judicial District Court')
        data_pieces.remove('Primary Sort:')
        data_pieces.remove('Case Number')
        data_pieces.remove('Case Number')
        data_pieces.remove('Date Range Type:')
        data_pieces.remove('Case Filed Date')
        data_pieces.remove('Location')
        data_pieces.remove('Judicial Officer')
        data_pieces.remove('Case\nCategory')
        data_pieces.remove('Case Type\nCase Subtype')
        data_pieces.remove('Case Status')
        data_pieces.remove('Style')
        
        #Now iterate through the list, and insert each type of value into a specific list of strings representing each column
        #It looks like every piece of information flows in the right order, even though the pieces of information are mixed up

        cause_nums = []
        officers = []
        case_types = []
        styles = []
        statuses = []

        for piece in data_pieces:
            #Check if it matches cause number pattern
            if re.search(r'[0-9]{2}-[DCVR]{3}-[0-9A-Z]*', piece) != None:
                cause_nums.append(piece)
                continue

            #Check if part is the officer's name
            if piece == 'Krenek, Edward\nM.':
                officers.append(piece)
                continue

            #Check if part is the case status
            if piece in possible_statuses:
                statuses.append(piece)
                continue
                
            #Check if part is the case type
            if piece in possible_types:
                case_types.append(piece)
                continue

            #Check if part is the 'style' column
            #It should contain some variation of 'vs'
            if re.search(r'\s{1}vs\s{1}', piece.lower()) != None:
                styles.append(piece)
                continue
            elif re.search(r'\s{1}vs\.\s{1}', piece.lower()) != None:
                styles.append(piece)
                continue
            elif re.search(r'\s{1}v\s{1}', piece.lower()) != None:
                styles.append(piece)
                continue
            elif re.search(r'\s{1}v\.', piece.lower()) != None:
                styles.append(piece)
                continue
            elif re.search(r'in\s{1}re:\s{1}', piece.lower()) != None:
                styles.append(piece)
                continue

            #For now, assume that all this part can be at this point is the case type
            #Later, I will create a list of possible case types and verify the strings exists in that list
            case_types.append(piece)
        
        #For error checking, verify that each list is the same length. They should all have the same number of values
        num_cases = len(cause_nums)
        page_num = page_num + 1 #Plus one because page_num starts at zero

        if num_cases != len(officers):
            print(f"Error on Page {page_num}!")
            print('Num Cases Does Not Equal Number of Officer Values')
            break

        if num_cases != len(case_types):
            print(f"Error on Page {page_num}!")
            print('Num Cases Does Not Equal Number of Case Type Values')
            break

        if num_cases != len(styles):
            print(f"Error on Page {page_num}!")
            print('Num Cases Does Not Equal Number of Style Values')
            break

        if num_cases != len(statuses):
            print(f"Error on Page {page_num}!")
            print('Num Cases Does Not Equal Number of Status Values')
            break
            
        #Create a dictionary. Make the keys the names of the columns, and the values the lists.
        my_dict = {
            'Cause Number': cause_nums,
            'Officer': officers,
            'Case Type': case_types,
            'Style': styles,
            'Status': statuses
        }
        
        page_df = pd.DataFrame(my_dict)
        
        #Once we have the current page's dataframe, add it to the overall dataframe
        df = pd.concat([df,page_df], ignore_index = True)
        
    return df
```
